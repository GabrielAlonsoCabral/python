```python
#criando um jeito pra saber todos os paises
paises0=[]


for x in range(129971):
    paises0.append(df['country'][x])

paises1=[]
for j in paises0:
    if j not in paises1:
        paises1.append(j)
print('Quantidade de paises:',len(paises1))
print('lista dos paises:',paises1)
```

```python
France=[]
for k in paises0:

    if k =='France':
        France.append(k)
print(len(France))
```
