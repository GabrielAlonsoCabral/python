```python
frutasj={}
i=int(input('Qnt de tipos de Frutas:'))
for i in range(i):
    nome_da_fruta=input('Nome da Fruta:')
    quantidade_fruta=input('Quantidade de %s:'%nome_da_fruta)
    frutasj.update({nome_da_fruta:[quantidade_fruta,10]})
df=pd.DataFrame(frutasj,index=['Row1','Row2'])
print(df)
```
