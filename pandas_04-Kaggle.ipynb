```python
#aplicar a altera_pontos na tabela.
def remean_points(row):
    row.points = row.points - df.points.mean()

    return row

df1=df.apply(remean_points, axis='columns')
```

```python
def reprice(row):
    row.price=row.price*5.36
    return row
df2=df1.apply(reprice, axis='columns')
```
